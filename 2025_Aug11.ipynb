```python
nums = [3, 4, "hello", 5+6j, 6, 7]

def apply(fn, data):
    result = []
    for item in data:
        if type(item) in (int, float):
            r = fn(item)
        else:
            r = 0
        result.append(r)
    return result

apply(square, nums)
```

```python
users = ["john", "janes", 56, "mary", "eric", "claire"]

def to_upper(x): return x.upper()


def apply(fn, data):
    from numbers import Number
    result = []
    for item in data:
        if isinstance(item, Number):
            r = fn(item)
        else:
            r = 0
        result.append(r)
    return result

apply(to_upper, users)
```

```python
users = ["john", "janes", 56, "mary", "eric", "claire"]

def to_upper(x): return x.upper()


def apply(fn, data):
    result = []
    for item in data:
        if hasattr(item, "__mul__"):
            r = fn(item)
        else:
            r = 0
        result.append(r)
    return result

apply(to_upper, users)
```

```python
users = ["john", "janes", 56, "mary", "eric", "claire"]

def to_upper(x): return x.upper()


def apply(fn, data, default=None):
    result = []
    for item in data:
        try:
            r = fn(item)
        except:
            r = default
        result.append(r)
    return result

apply(to_upper, users, "invalid")
```

```python
s = "this is a test is a test string with repeated test words test times"

def count_words(data):
    count = {}
    for word in data.split():
        #if word not in count:
        #    count[word] = 0
        #count[word] += 1
        count[word] = count.get(word, 0) + 1
        
    for w, c in count.items():
        print(f"{w}: {c} times")

count_words(s)
```

```python
s = "this is a test is a test string with repeated test words test times"

from collections import defaultdict
def count_words(data):
    count = defaultdict(int)

    for word in data.split():
        count[word] += 1
            
    for w, c in count.items():
        print(f"{w}: {c} times")

count_words(s)
```

```python
s = "this is a test is a test string with repeated test words test times"

from collections import Counter
def count_words(data):
    count = Counter(data.split())
            
    for w, c in count.items():
        print(f"{w}: {c} times")

count_words(s)
```
